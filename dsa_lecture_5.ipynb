```python
''' 

given a, N, p
implement (a**N) % p

remember:
    ( a*b ) % m = ( a%m * b%m ) % m

'''
def get_pow(a, N, p):
    if a == 1:
        return 1
    if N == 1:
        return a % p
    
    return get_pow(a, N-1, p) * ( (a%p) * (a%p) % p )

a, N, p = map(int, list(input()))
print(get_pow(a, N, p))
```

```python
# Solution 2 -> time complexity is O(N)
def get_pow(a, N, p):

    if a == 1 or N == 0:
        return 1
    
    return ( (a%p) * get_pow(a, N-1, p) ) % p

a, N, p = map(int, list(input()))
print(get_pow(a, N, p))
```

```python
# Solution 3 -> time complexity is O(N)

def get_pow(a, N, p):
    if ( N==0 or a ==1 ):
        return 1
    if ( N%2==0 ):
        return ( get_pow(a, N//2, p) * get_pow(a, N//2, p) ) % p 
    else:
        return ( get_pow(a, N//2, p) * get_pow(a, N//2, p)  * (a % p) ) % p 
```

```python
# Solution 4 -> Optimized Solution -> Time Complexity logn

def get_pow(a, N, p):
    
    if ( N==0 or a ==1 ):
        return 1
    
    hp = get_pow(a, N//2, p)
    
    if ( N % 2 == 0 ):
        return ( hp * hp ) % p
    
    else:
        return (( hp * hp ) * (a%p) ) % p

''' 
(( hp * hp ) * (a%p) ) % p
=> ((( hp * hp ) % p ) * ((a%b) % p) ) % p
=> ((( hp * hp ) % p ) * (a%b) ) % p
'''
    
```

```python
''' 
Given Array print the array using Recursion
'''

def printf( arr, i, N ):
    if i == N:
        return
    print(arr[i], end=",")
    printf( arr, i+1, N )

arr = list(map(int, list(input())))
printf(arr, 0, len(arr))
```

```python
''' 
Solve nCr give nCr = (n-1)C(r-1) + (n-1)C(r)
'''

def func(n, r):
    if ( r==0 or n==r ) :
        return 1
    return func(n-1, r-1) + func(n-1, r)

n, r = map(int, list(input()))
print(func(n, r))
```
