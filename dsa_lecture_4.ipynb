```python
# Sum of N number

def get_sum(n):
    if n==1:
        return 1
    return n + get_sum(n-1)

n = int(input())
print(get_sum(n))
```

```python
# Factorial of a number

def get_fact(n):
    if n == 1:
        return 1
    return get_fact(n-1) * n 

n = int(input())
print(get_fact(n))
```

```python
''' 
Palindrome : whether number is palindrome or not
'''
in_str = input()

def check_palindrome(in_str, prev, next):
    if prev==next:
        return 1
    if in_str[prev] != in_str[next]:
        return 0
    return check_palindrome(in_str, prev+1, next-1)

print("Palindrome") if check_palindrome(in_str, 0, len(in_str)-1) else print("Not a Palindrome")
```

```python
''' 
Count Digit
'''
def count_digit(n):
    if n == 0:
        return 1
    return 1 + count_digit(n//10)

print(count_digit(int(input())))
```

```python
''' 
sum of digits
'''
def sum_digits(n):
    if n == 0:
        return 0
    return n % 10 + sum_digits( n // 10 )

print(sum_digits(int(input())))
```
