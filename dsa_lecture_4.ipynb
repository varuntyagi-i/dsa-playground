### Recursion

```python
# Sum of N number

def get_sum(n):
    if n==1:
        return 1
    return n + get_sum(n-1)

n = int(input())
print(get_sum(n))
```

```python
# Factorial of a number

def get_fact(n):
    if n == 1:
        return 1
    return get_fact(n-1) * n 

n = int(input())
print(get_fact(n))
```
